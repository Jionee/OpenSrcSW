# 실습1 - 서울의 일교차

```python
# 서울에서 일교차가 가장 큰 날, 최저 기온이 가장 낮은 날, 최고 기온이 가장 높은 날을 찾아 출력하시오
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data) #header 건너띄기 위함

#일교차 : 최고 - 최저 (row[4] - row[3])
#최저 기온 : row[3]
#최고 기온 : row[4]

max_dif = 0
min_low = 999
max_high = 0
date = ['']*3

for row in data:
    for i in range(2,5):
        dif = float(row[4]) - float(row[3])
        low = float(row[3])
        high = float(row[4])
        
        if dif > max_dif:
            max_dif = dif
            date[0] = row[0]
        if high > max_high:
            max_high = high
            date[1] = row[0]
        if low < min_low:
            min_low = low
            date[2] = row[0]
        
print("일교차가 가장 큰 날 : ", date[0])
print("최저기온이 가장 낮은 날 : ", date[2])
print("최고기온이 가장 높은 날 : ", date[1])
        
```

# 실습2 - 지하철 이용객

```python
#노선별로 가장 이용객이 많은 3개역, 가장 적은 3개 역 
#이용객 : 승차+하차
import csv
import re
import operator
f = open('subwayUsers2.csv','r',encoding = "utf-8")
data = csv.reader(f)
next(data)

def removeComma(a):
    char = a #940,925 > 940925로 변경
    p = re.compile("[^,]")
    a = "".join(p.findall(char))
    a = int(a)
    return a

dic = dict()
for row in data:
    subway = [row[3],removeComma(row[4])+removeComma(row[5])]
    if row[1] in dic:
        dic[row[1]].append(subway)
    else:
        dic[row[1]] = [subway]


for key,value in dic.items():
    print("========="+key+"==========")
    value = sorted(value, key = lambda x : x[1])
    #print(value)
    print("@ 이용객이 가장 많은 역 순위 @")
    for i in range(-1,-4,-1):
        print(value[i][0],value[i][1])
    print("@ 이용객이 가장 적은 역 순위 @")
    for i in range(0,3):
        print(value[i][0],value[i][1])
```

```python
#서울의 최고 기온이 가장 높았던 날은 언제였고, 몇 도였을까?
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data) #header 건너띄기 위함
max_temp= -999
max_data = ''
for row in data:
    if row[-1] == '': #최고 기온 데이터 누락 시
        row[-1] = -999
        
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_data = row[0]
        max_temp = row[-1]
f.close()
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은', max_data+'로, ',max_temp,'도 였습니다.')
```

# 수업 (대중교통)

```python
#유임/무임 최대 비율
import csv
import re
f = open('subwayfee.csv','r',encoding = "utf-8")
data = csv.reader(f)
next(data)
mx = 0
rate = 0
for row in data:
    for i in range(4,8):
        char = row[i] #940,925 > 940925로 변경
        p = re.compile("[^,]")
        row[i] = "".join(p.findall(char))
        
        row[i] = int(row[i])
    if row[6] != 0:
        rate = row[4] / row[6]
        if rate > mx:
            mx = rate
print(mx)
```

```python
#유임 승차 비율이 가장 높은 역
#유임/(유임+무임)
import csv
f = open('subwayfee.csv','r',encoding = "utf-8")
data = csv.reader(f)
next(data)

mx = 0
rate = 0
mx_station = ''

for row in data:
    for i in range(4,8):
        char = row[i] #940,925 > 940925로 변경
        p = re.compile("[^,]")
        row[i] = "".join(p.findall(char))
        
        row[i] = int(row[i])
        
    if (row[4]+row[6]) != 0:
        rate = row[4] / (row[4] + row[6])
        if rate > mx:
            mx = rate
            mx_station = row[3] + ' ' + row[1]
            
print(mx_station, round(mx * 100,2))
```

```python
#유무임 승하차 인원이 가장 많은 역
import csv
f = open('subwayfee.csv','r',encoding = "utf-8")
data = csv.reader(f)
next(data)

mx = [0]*4
mx_station = ['']*4
label = ['유임승차','유임하차','무임승차','무임하차']

for row in data:
    for i in range(4,8):
        char = row[i] #940,925 > 940925로 변경
        p = re.compile("[^,]")
        row[i] = "".join(p.findall(char))
        
        row[i] = int(row[i])
        if row[i] > mx[i-4]:
            mx[i-4] = row[i]
            mx_station[i-4] = row[3] + ' ' + row[1]
for i in range(4):
    print(label[i] + ' : ' + mx_station[i] , mx[i])
```
