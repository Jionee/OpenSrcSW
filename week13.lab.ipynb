# 실습1

```python
#실습1 : 생일 기온 시각화
#자신 생일의 서울 지역 최고 기온과 최저 기온을 아래와 같이 시각화하시오.

import csv
import matplotlib.pyplot as plt
f = open('seoul2.csv')
data = csv.reader(f)
next(data)

result = [[],[]]
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '07':
            result[0].append(float(row[-1]))
            result[1].append(float(row[-2]))
print(result)

plt.title('My birthday')
plt.plot(result[0],color = 'deeppink',label='최고기온')
plt.plot(result[1],label='최저기온')
plt.legend()
plt.show()
```

# 실습2

```python
#실습2 : 최고 기온 시각화
#서울 지역의 1월과 8월 최고 기온 데이터를 히스토그램으로 시각화하시오

import csv
import matplotlib.pyplot as plt
f = open('seoul2.csv')
data = csv.reader(f)
next(data)

jan = []
aug = []
for row in data:
    if row[0].split('-')[1] == '01': #1월 최고기온
        if row[-1] != '':
            jan.append(float(row[-1]))
    elif row[0].split('-')[1] == '08': #8월 최고기온
        if row[-1] != '':
            aug.append(float(row[-1]))
#print(jan,aug)

plt.hist(jan,bins=100,label='Jan')
plt.hist(aug,bins = 100,label='Aug')
plt.legend()
plt.show()
```

```python
#내 생일 서울의 최고 기온 시각화
import csv
f = open('seoul2.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '07':
            result.append(float(row[-1]))
print(result)
```

```python
import matplotlib.pyplot as plt
plt.title('My birthday')
plt.plot(result)
plt.show()
```

```python
#히스토그램
plt.hist(result,color='r')
plt.show()
```

```python
#상자그래프
plt.boxplot(result)
plt.show()
```

```python
#1월부터 12월까지 최고 데이터 > 상자 그래프
f = open('seoul2.csv')
data = csv.reader(f)
next(data)

#각 []안에는 n월의 1,2...31일까지의 최고기온이 들어가 있음
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
next(data)

m = []
f = []
for row in data:
    if '동작구' in row[0].split(' ')[1]:
        print(row)
        for i in row[3:14]:
            m.append(-(int(i.replace(',',''))))
        for i in row[16:]:
            f.append(int(i.replace(',','')))
print(m,f)

#항아리모양
plt.rc('font',family='Malgun Gothic')
plt.title('동작구의 남녀 성별 인구 분포')
plt.barh(range(11),m,label='남성')
plt.barh(range(11),f,label='여성')
plt.legend()
plt.show()
```
